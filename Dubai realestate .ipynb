```python
tot_agg_dataset =Transactions.groupby(['area_name_en']).max()# gives max for every area's everything
#proc_area=tot_agg_dataset['actual_worth']
#proc_area=proc_area.nlargest(10, ['actual_worth'])
# gives max price sold for 
a=Transactions['actual_worth'].groupby(Transactions['area_name_en']).max()  #or use .nlargest(1)
# Takes 10 largest after grouping and displays the 10 largest grouped value  a.nlargest(10) , Grouped by area therefore Index is area 
a.nlargest(10).plot.bar(figsize=(20,5), stacked=True , color=['red', 'green'],y='actual_worth')
plt.title(" Top 10 Areas with the Highest Valued Property ")
plt.xlabel("Name of areas ")
plt.ylabel("The price Properties were stored for ")
```

```python
a=Transactions['actual_worth'].groupby(Transactions['area_name_en']).sum()  #or use .nlargest(1)
# Takes 10 largest after grouping and displays the 10 largest grouped value  a.nlargest(10) , Grouped by area therefore Index is area 
a.nlargest(10).plot.barh(figsize=(20,5), stacked=True , color=['red', 'green'],y='actual_worth')
plt.title(" Areas with the Most amount of of properties sold Overtime  ")
plt.xlabel("Name of areas ")
plt.ylabel("The price Properties were stored for ")
# To check it properly check the 10 most Value count of areas(amount of unit stored )and check the most sum of property sold FIND COORELATION .
```

```python
a2=Transactions['actual_worth'].groupby(Transactions['area_name_en']).median()  #or use .nlargest(1)
# Takes 10 largest after grouping and displays the 10 largest grouped value  a.nlargest(10) , Grouped by area therefore Index is area 
a2.nlargest(10).plot.barh(figsize=(12,5), stacked=True , color=['pink', 'green'],y='actual_worth')
a2.head(10)
plt.title(" Areas with Most median Propery Sold Overtime ")
plt.xlabel("Name of areas ")
plt.ylabel("The price Properties were stored for ")
plt.legend()
tot_agg_dataset12 =Transactions.groupby(['area_name_en']).median()# gives max for every area's everything
tot_agg_dataset12.nlargest(10,"actual_worth")# can do plot.bar directly on this 
```
